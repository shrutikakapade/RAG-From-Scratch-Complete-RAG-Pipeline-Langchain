## Document Processing Pipeline

In this notebook, we implement the core steps required to prepare documents for a **RAG pipeline**.

- **Document Loading:** Load the source document into the system.
- **Text Splitting:** Split the document into smaller chunks for efficient processing and retrieval.
- **Embedding Generation:** Convert each text chunk into vector embeddings that capture the semantic meaning of the text.

These embeddings can then be stored in a **vector database** to enable fast and accurate semantic search.

In this step, we create a vector database using LangChain and Chroma.
The purpose of this database is to store document chunks along with their embeddings so that we can perform semantic search during retrieval.

```python
#Creating Database

database = Chroma.from_documents(
    documents=final_text_split[:10],
    embedding=emb_llm,
    persist_directory="chroma_db",
    collection_name="Data_Analysis_interview_questions"
)

# Note : persist_directory should be a folder, not a .db file.
# Chroma automatically creates the database files inside the folder.
# When Google Generative AI embeddings process many chunks at once, the API may fail.
```

#### Chroma.from_documents()

This method creates a vector database directly from a list of documents.
It performs three main tasks automatically:

- Converts documents into embeddings

- Stores the embeddings in the vector database

- Saves the database locally for future use

### documents=final_text_split[:10]

- final_text_split contains the document chunks created using a text splitter.

- [:10] means only the first 10 chunks are used.

- This is helpful because embedding too many chunks at once can sometimes cause API failures or slow processing.

#### embedding=emb_llm

- This specifies the embedding model used to convert text into numerical vectors.

- The embedding model is created using Google Generative AI embeddings, which transform text into vector representations that capture semantic meaning.

#### persist_directory="chroma_db"

- This tells Chroma to save the database locally in a folder named chroma_db.

- Important points:It must be a folder, not a .db file.

- Chroma automatically creates the required files inside the folder.

#### collection_name="Data_Analysis_interview_questions"

- A collection is like a table inside the vector database.

- Here, the collection stores embeddings related to Data Analysis interview questions.

```python
from re import search
vec = database.as_retriever(
    search_type="similarity",
    search_kwargs={"k":2})
```

#### database.as_retriever()
- Converts the vector database into a retriever that can search and return relevant document chunks.

#### search_type="similarity"
- Uses semantic similarity search to find documents whose embeddings are most similar to the query embedding.

#### search_kwargs={"k": 2}
- Returns the top 2 most relevant document chunks from the database.

```python
Document_objects =vec.invoke("What is data analysis")

# When a query is executed:
# The retriever returns a list of Document objects
```

### RAG Retrieval Flow

In a Retrieval-Augmented Generation (RAG) system, the process of generating an answer follows these steps:


**Explanation:**

1. **Retriever**  
   Searches the vector database and finds the most relevant document chunks related to the user query.

2. **page_content**  
   The retrieved documents contain text in the `page_content` field. This is the **actual information extracted from the source document**.

3. **LLM**  
   The retrieved `page_content` is passed to the Large Language Model, which reads the context and understands the information.

4. **Final Answer**  
   The LLM generates a clear and contextual response based on the retrieved document content.
