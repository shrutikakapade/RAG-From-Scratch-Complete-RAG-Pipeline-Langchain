```python
# step 2 : Text_Splitter --> Convert the text into small samll chunks

from langchain_text_splitters import RecursiveCharacterTextSplitter

rts = RecursiveCharacterTextSplitter(separators= ["\n\n","\n"," "],
                                     chunk_size =500 , chunk_overlap =10)

# Seperated by "\n\n":Paregraph , "\n":Sentence, " ":Space ,"":word
#chunk_overlap = 10 means:
# When splitting text into chunks, the next chunk repeats the last 10 characters of the previous chunk.
```

```python
chunks = rts.split_documents(document)
```

```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
emd = HuggingFaceEmbeddings(model = "all-MiniLM-L6-v2")
```

```python
vec_db = Chroma.from_documents(documents =chunks, embedding= emd, persist_directory="db")
vec_db.persist()
```

```python
# Try to make database Retrieval

retriever = vec_db.as_retriever(search_kwargs={"k":2})#This is going to wrap vector database as a runnable object called retriever
```

```python
retriever.invoke("Hi how are you")
```

```python
len(retriever.invoke("Hi how are you")) #search_kwargs={"k":2}

# retriever it is a runnable
```

```python
# Function: Combines two outputs into a single, well-structured input for the LLM using ChatPromptTemplate, ensuring better context and response quality.
def f1(data):
  l = []

  for y in data: #Loop through documents
    l.append(y.page_content)
  return "\n\n".join(l)
```

```python
from langchain_core.runnables import RunnableLambda, RunnableParallel, RunnablePassthrough, RunnableSequence
```

```python
runn2 = RunnableLambda(f1)
```

```python
RunnableSequence(retriever , runn2).invoke("Hi,how are you")

# runn2 is a function which get input(data) from retriever one by one and combine it into one output.
# retriever is also a runnable which is able to retrieved the relevant information from the vector database using similarity search approach.

# So this is our first chain which contain output
```

```python
chain1 = RunnableSequence(retriever,runn2)
```

```python
chain1.invoke("Hi,how are you") # we get one output
```

```python
RunnableParallel({"context":chain1, "question": RunnablePassthrough()}).invoke("Hi,how are you")

# Context + Question = Better LLM Response Quality.
```

```python
chain2 = RunnableParallel({"context":chain1, "question": RunnablePassthrough()})
```

```python
from langchain_core.prompts import ChatPromptTemplate, HumanMessagePromptTemplate, SystemMessagePromptTemplate

#Runnable 1
cpt = ChatPromptTemplate.from_messages([SystemMessagePromptTemplate.from_template("You are a helpful HR assistant who is having 4 years of experience"),
                                       HumanMessagePromptTemplate.from_template("Answer the question based on the below given context and if there is no context then please return I don not know {context}\n\n{question}")])
```

# Fallback RAG (Retrieval-Augmented Generation)

## What is Fallback RAG?
Fallback RAG is an enhanced version of the standard Retrieval-Augmented Generation (RAG) approach.  
In this method, when the primary retrieval system fails to fetch relevant or sufficient context, a **fallback mechanism** is triggered.

This fallback can involve:
- Querying an alternative knowledge source
- Expanding or reformulating the query
- Using a broader or secondary vector database
- Allowing the LLM to answer with general knowledge (as a last resort)

👉 In simple terms:  
**If the first retrieval step fails, the system “falls back” to another strategy to still provide a useful answer.**

---

## Why is Fallback RAG Used?
Fallback RAG is used to improve the **robustness and reliability** of AI systems.

### Key Reasons:
- To handle **missing or incomplete data**
- To reduce **empty or irrelevant responses**
- To improve **user experience** by always providing an answer
- To make systems more **fault-tolerant in real-world scenarios**

---

```python
# Create a fallBack function

def fallback(data):
  answer = main_chain.invoke(data)

  if answer == "False":
    print("now i am calling the retriever")
    return final_RAG_Pipeline.invoke(data["question"])
  else:
    print("now i am using LLM")
    return answer
```

```python
fallback_rag = RunnableLambda(fallback)
```

```python
fallback_rag.invoke({"question":"In my company how many leaves are available"})
```

```python
fallback_rag.invoke({"question":"Tell me about Deep Learning"})
```

```python
fallback_rag.invoke({"question":"Tell me about the polices in SIRCA PAINTS INDIA LTD NEW DELHI"})
```



## How It Works (Simple Flow)
1. User query is received  
2. Primary retriever searches for relevant documents  
3. If relevant context is found → Generate answer  
4. If not found → Trigger fallback:
   - Retry retrieval (query reformulation)
   - Use secondary database / API
   - Use LLM knowledge as backup  
5. Return the best possible response  

---

## Advantages of Fallback RAG

### 1. Improved Reliability
- Ensures the system rarely fails to respond

### 2. Better User Experience
- Users get answers even when data is missing

### 3. Reduced Failure Cases
- Handles edge cases where retrieval is weak

### 4. Flexibility
- Can integrate multiple data sources and strategies

### 5. Increased Coverage
- Expands the system’s ability to answer diverse queries

---

## Disadvantages of Fallback RAG

### 1. Risk of Hallucination
- If fallback uses LLM knowledge, answers may be less accurate

### 2. Increased Complexity
- More components → harder to design and maintain

### 3. Higher Latency
- Additional retrieval steps can slow down response time

### 4. Cost Overhead
- More API calls and computations increase cost

### 5. Inconsistent Responses
- Answers may vary depending on which fallback path is used

---

## Final Insight
Fallback RAG is a **practical enhancement** to standard RAG systems.  
It prioritizes **availability over perfection**, ensuring that users always receive a response—even if the system must rely on alternative strategies.

👉 Best Practice:  
Use fallback mechanisms carefully and combine them with **confidence scoring** or **answer validation** to maintain accuracy.
