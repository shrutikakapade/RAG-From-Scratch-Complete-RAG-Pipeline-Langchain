# *With using runnable (New version: mainly used in industry(real world application))*

```python
from langchain_core.runnables.history import RunnableWithMessageHistory # Wrapper
from langchain_core.chat_history import InMemoryChatMessageHistory #RedisChatMessageHistory #All the different type of session_store
from langchain_core.runnables import RunnablePassthrough
# InMemoryChatMessageHistory: It is called local memory, when the sessional memory is over all the chat will be losed
# RedisChatMessageHistory: It is a CLOUD PLATFORM by using which we can create a cloud database
# Redis is a platform : Data stored in RAM → microsecond-level latency
# Much faster than disk-based DBs like MySQL
# Unlike traditional databases, Redis stores data in RAM, making it extremely fast
```

```python
# Summary chain

#Create a ChatpromptTemplate (cpt)

cpt = ChatPromptTemplate.from_messages([SystemMessagePromptTemplate.from_template("Please summarize the past conversations {history}")])

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
parser = StrOutputParser()

summary_chain = cpt|model|parser
```

```python
# The summary_chain could by the function
# function get data as a dictionary
# Function to make it as a router

def router(data):
  if data["history"] and len(data["history"])>0:
    return summary_chain.invoke(data)
  else:
    return "No history found"
```

```python
def router(data):
  if data["history"] and len(data["history"])>10:
    return summary_chain.invoke(data)
  else:
    return data

  # Here we save the tokens
```

```python
# Create a session_factory and session_store

storage ={} #session store

def session_factory(session_id):
  if session_id not in storage:
   storage[session_id] = InMemoryChatMessageHistory()
  return storage[session_id]
```

```python
# Create a main chain where we actually give our query

#Create a ChatpromptTemplate (cpt)
# Summary Chain
cpt = ChatPromptTemplate.from_messages([SystemMessagePromptTemplate.from_template("You are a helpful assistance and you are given the past data as summary:{summary}"),
                                        HumanMessagePromptTemplate.from_template("{input}")])

# when use ConversationSummaryBufferMemory: we use parameter called MessagesPlaceholder
# MessagesPlaceholder: It is also prompt template by using which we can insert a list of message inside your message (ChatPromptTemplate.from_messages).


model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
sto = StrOutputParser()
```

```python
main_chain =RunnablePassthrough.assign(summary=router)|cpt|model|sto #main chain
```

```python
app = RunnableWithMessageHistory(main_chain,session_factory,history_messages_key="history",input_messages_key="input")
```

*1. ConversationSummaryMemory (Manual Summary Version)*

```python

from langchain_classic.memory import ConversationSummaryMemory

memory = ConversationSummaryMemory(llm=model)

def get_summary_history(session_id: str):
    if session_id not in storage:
        storage[session_id] = {
            "history": InMemoryChatMessageHistory(),
            "memory": memory
        }
    return storage[session_id]["history"]

with_message_history = RunnableWithMessageHistory(
    main_chain,
    get_summary_history,
    input_messages_key="input",
    history_messages_key="history"
)

# Example call
response = with_message_history.invoke(
    {"input": "Hi, I am Shrutika"},
    config={"configurable": {"session_id": "1"}}
)

print(response)
```

*2. ConversationSummaryBufferMemory*

```python

from langchain_classic.memory import ConversationSummaryBufferMemory

memory = ConversationSummaryBufferMemory(
    llm=model,
    max_token_limit=100
)

def get_buffer_history(session_id: str):
    if session_id not in storage:
        storage[session_id] = {
            "history": InMemoryChatMessageHistory(),
            "memory": memory
        }
    return storage[session_id]["history"]

with_message_history = RunnableWithMessageHistory(
    main_chain,
    get_buffer_history,
    input_messages_key="input",
    history_messages_key="history"
)

response = with_message_history.invoke(
    {"input": "I am learning AI"},
    config={"configurable": {"session_id": "1"}}
)

print(response)
```

*3. ConversationSummaryBufferWindowMemory (Window + Summary)*

```python
from langchain_community.memory import ConversationSummaryBufferWindowMemory

memory = ConversationSummaryBufferWindowMemory(
    llm=model,
    k=3
)

def get_window_history(session_id: str):
    if session_id not in storage:
        storage[session_id] = {
            "history": InMemoryChatMessageHistory(),
            "memory": memory
        }
    return storage[session_id]["history"]

with_message_history = RunnableWithMessageHistory(
    main_chain,
    get_window_history,
    input_messages_key="input",
    history_messages_key="history"
)

response = with_message_history.invoke(
    {"input": "What do you remember about me?"},
    config={"configurable": {"session_id": "1"}}
)

print(response)
```

### 1. ConversationSummaryMemory
- Stores only summary of the conversation.
- Old messages are converted into short summary.
- Does NOT keep exact chat history.
- Use when: long conversations, low token cost.

### 2. ConversationSummaryBufferMemory
- Stores: Recent messages (buffer), Summary of older messages.
- When limit exceeds → older chats get summarized
- Use when: need balance of context + efficiency

### 3. ConversationSummaryBufferWindowMemory
- Stores: Last K messages (fixed window), Summary of older messages.
- Window size is fixed (like last 3–5 chats)
- Use when: chatbot needs strong recent context

*SummaryMemory stores only summarized chat, BufferMemory keeps recent chats plus summary, and BufferWindowMemory keeps last K messages with summary for better recent context.*
