### What is a Text Splitter?

- In a Retrieval-Augmented Generation pipeline, a text splitter is a tool that breaks large documents into smaller chunks of text so that AI models can process and understand them better.

### Why Do We Need a Text Splitter?

- Large Language Models (LLMs) like GPT-4 or Gemini cannot handle extremely long documents at once.
So we split the document into smaller pieces (chunks) and store them in a vector database for retrieval.

### 1. Text_Splitter : CharacterTextSplitter()

- CharacterTextSplitter is a text splitting tool in LangChain used to divide large text into smaller chunks based on the number of characters.

- Large documents cannot be processed directly by Large Language Models because of token limits. So we split the text into smaller parts called chunks before using it in RAG pipelines.

```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
text_split = CharacterTextSplitter()
```

```python
text_split.split_text(doc) #It is converted txt document into chunks

# Note: We use split_text when the data is in plain text format. If the data is in document format (like LangChain Document objects), we use split_documents instead.
```

```python
# Check the count of chunk
len(text_split.split_text(doc))

# The chunk size is : 3
```

```python
text_split.split_text(doc)[0]
```

```python
text_split.split_text(doc)[1]
```

```python
text_split.split_text(doc)[2]
```

Note: Here we can see that each chunk has a large size, which we need to reduce. Smaller chunks help the model retrieve more precise and relevant information, improve embedding quality, and ensure the LLM can process the context efficiently without exceeding token limits.

```python
# How to reduce the chunk size:
text_split = CharacterTextSplitter(
    chunk_size=500 ,
    separator=" ",
    chunk_overlap = 10)

# Note : With CharacterTextSplitter, we can specify only one separator (for example " "). However, RecursiveCharacterTextSplitter allows us to use multiple separators, such as paragraphs, lines, and words, enabling the text to be split more intelligently while preserving its structure and context.
```

Overlap means repeating a small part of the previous chunk in the next chunk.

Here, 10 characters (or tokens depending on the splitter) from the end of the previous chunk will also appear at the start of the next chunk.

```python
len(text_split.split_text(doc))
```

```python
text_split.split_text(doc)[0]
```

```python
text_split.split_text(doc)[1]
```

Doc[0] contains the first chunk of the text, while Doc[1] begins with a few words that already appeared at the end of Doc[0]. This occurs due to chunk overlap, which intentionally repeats a small portion of the previous chunk to preserve context and ensure continuity of information between chunks.

This is useful because it prevents important information from being lost during splitting and improves the accuracy of information retrieval in RAG systems.

### Note: The main disadvantage of CharacterTextSplitter is that it splits text purely based on character length, which can break sentences and reduce contextual meaning in chunks.

### 2. Text_Splitter : RecursiveCharacterTextSplitter()

We switch to RecursiveCharacterTextSplitter in LangChain because CharacterTextSplitter may break sentences in the middle.

RecursiveCharacterTextSplitter tries to split text more carefully by first checking paragraphs, then lines, then words, so the meaning of the text is preserved.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_split_2 = RecursiveCharacterTextSplitter(
    chunk_size=400 ,
    separators=[" ","\n","e","\n\n"],
    chunk_overlap = 30)
```

```python
len(text_split_2.split_text(doc))
```

```python
len(text_split_2.split_text(doc)[0])
```

```python
len(text_split_2.split_text(doc)[21])
```

RecursiveCharacterTextSplitter intelligently splits text using multiple separators, ensuring chunks preserve meaning, context, and structure while staying within the defined size limit.

### 3. Text_Splitter : CodeTextSplitter()

- CodeTextSplitter is used to split programming code, but due to its limited flexibility we often use RecursiveCharacterTextSplitter with a [language parameter], which provides language-aware code splitting while remaining more adaptable for different data types.


#### The language parameter tells the splitter which programming language rules to follow.

Example languages:python, javascript, java, cpp, go, html

*Using this parameter allows the splitter to:

- Understand code structure

- Split at logical code boundaries

- Maintain function and class integrity

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter , Language
```

```python
doc = '''
def fun_1(n,m):
  print(n+m)
  print(n-m)

fun_1(10,20)

def fun_2(n,m):
  print(n*m)
  print(n/m)

fun_2(5,6)
fun_2(7,8) '''

#It is a python language
```

```python
code_split = RecursiveCharacterTextSplitter.from_language(
    language = Language.PYTHON,
    chunk_size=30,
    chunk_overlap = 2)
```

```python
code_split.split_text(doc)
```

```python
len(code_split.split_text(doc))
```

### Way 4: Transformers

- Transformers do not split text directly.
- Instead, they provide tokenizers that convert text into tokens.

'' Example tokenizer class:AutoTokenizer ''

These tokens are then used by text splitters.

First, we install the Hugging Face Transformers library because it provides tokenizers that convert text into tokens. Large Language Models process text in tokens rather than characters, so using a tokenizer helps us split text more accurately according to the model’s input format.

```python
!pip install transformers
```

```python
from transformers import AutoTokenizer
# AutoTokenizer automatically loads the correct tokenizer for a specific model.
```

```python
tokens = AutoTokenizer.from_pretrained("deepseek-ai/deepseek-coder-1.3b-instruct")
# Here, the tokenizer understands how the DeepSeek-Coder model converts text into tokens.
```

```python
# Now, instead of splitting text by characters, we create a text splitter that splits text based on tokens.
token_split = RecursiveCharacterTextSplitter.from_huggingface_tokenizer(
    tokenizer = tokens,
    chunk_size =200,
    chunk_overlap = 30)
```

tokenizer = tokens
This tells the text splitter to use the Hugging Face tokenizer to count tokens instead of characters.

chunk_size = 200
Each chunk will contain a maximum of 200 tokens.

chunk_overlap = 30
The next chunk will repeat 30 tokens from the previous chunk so that context is preserved.


```python
len(token_split.split_text(doc))
```

```python
len(token_split.split_text(doc)[0])
```


This approach is very useful in Retrieval-Augmented Generation systems because it ensures that chunks follow the actual token limits of the LLM, improves embedding quality, and prevents the model from exceeding its context window.

In simple terms, using a tokenizer allows us to split documents the same way the model reads them, which makes the RAG pipeline more efficient and reliable.
