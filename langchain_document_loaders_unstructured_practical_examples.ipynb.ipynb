### Problem When Using TextLoader for `.md` Files

- Markdown (`.md`) files contain structured elements such as:
  - `#` Headers (H1, H2, H3)
  - Titles
  - Sections
  - Bullet points
  - Lists
  - Code blocks

- Each of these elements has its own meaning and structure.

- When we use **TextLoader**:
  - It does not understand Markdown structure.
  - It does not preserve headings or sections.
  - It ignores formatting symbols like `#`, `**`, `-`, etc.

- It extracts the entire file as **plain text**.

- As a result:
  - Document hierarchy is lost.
  - Semantic meaning is reduced.
  - Chunking becomes less meaningful.
  - Retrieval quality in RAG systems may decrease.

#### Simple Conclusion

- TextLoader treats `.md` files the same as `.txt` files.
- It does not preserve Markdown formatting or structure.

TextLoader reads both .txt and .md files as plain text. The difference lies in file structure — .md contains formatting and semantic markers, but TextLoader does not interpret them. For structured Markdown parsing, specialized loaders like UnstructuredMarkdownLoader are preferred.

## Difference between UnstructuredURLLoader and WebBaseLoader

### 1. WebBaseLoader
- `WebBaseLoader` in `LangChain` is used to load content from a webpage.
- It extracts text from the URL and returns it as **one single document**.
- It is simple and useful when you only need the **entire webpage content as one block of text**.

### 2. UnstructuredURLLoader
- `UnstructuredURLLoader` loads webpage content using the **Unstructured library**.
- It provides **more control and better structure** while extracting webpage content.

### Modes in UnstructuredURLLoader

#### mode = "single"
- The whole webpage is returned as **one document**.
- In this case, it behaves **similar to WebBaseLoader**.

#### mode = "elements"
- The webpage is divided into **multiple elements** such as:
  - Headings
  - Paragraphs
  - Lists
  - Other structured sections
- This makes it **more effective**, because we can **access and process each element of the page separately and easily**.

## Difference between UnstructuredCSVLoader and CSVLoader

### 1. CSVLoader
- `CSVLoader` in `LangChain` is used to **load data from a CSV file**.
- Each **row of the CSV file becomes one document**.
- The content is usually stored as **plain text with column names and values**.
- It is **simple and useful for structured tabular data**.

### 2. UnstructuredCSVLoader
- `UnstructuredCSVLoader` loads CSV files using the **Unstructured library**.
- It focuses on **extracting content more flexibly**.
- It can **process CSV data as elements or structured content**.
- It is useful when working with **document-processing pipelines or advanced data extraction**.
