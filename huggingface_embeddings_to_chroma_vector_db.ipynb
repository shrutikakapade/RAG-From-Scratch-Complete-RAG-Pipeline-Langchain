```python
from langchain_core.documents import Document #create a dummy database
# langchain_core is the library, documents is a module, and Document is a class used to store text data along with metadata in LangChain.
```

```python
#create a dummy database
data = [Document(page_content="langchain is a framework to create llm based application",metadata={"topic":"Langchain","difficultly":"easy"}),
        Document(page_content="langchain contains four important components langchain, langgraph, langsmith, langserve",metadata={"topic":"Langchain","difficultly":"medium"}),
        Document(page_content="ML stands for machine Learning",metadata={"topic":"machine Learning","difficultly":"easy"}),
        Document(page_content="Machine Learning having 3 types of machine Learning",metadata={"topic":"Machine Learning","difficultly":"easy"})
        ]
```

```python
db = Chroma.from_documents(documents = data, embedding=emb,persist_directory="db1")
db.persist()

#Chroma.from_documents(...) : It creates a vector database from your documents
#documents = Document(page_content="langchain is a framework to create llm based application",metadata={"topic":"Langchain","difficultly":"easy"})
#embedding= emb [Converts text → numerical vectors]
#persist_directory = Where to save the database on disk
#db = your Chroma vector database
#db.persist() = Saves the database permanently
```

db1 contain 4 things
1. document
2. Vectors = Stores vector embeddings
3. ids = Stores indexing data
4. metadata

persist_directory="db1" stores the complete Chroma vector database, including embeddings, metadata, documents, and index files for fast retrieval.

#### Note : db.persist = Persistence in a vector database means storing embeddings and data permanently on disk so they can be reused later without recomputation.

- ❌ Non-persistent -> Data stored in RAM (temporary) -> Lost after runtime
- ✅ Persistent -> Data stored on disk (permanent) -> Saved in folder like db1

### Creating a Chroma database with custom IDs allows precise control over document identification, making it easier to manage, update, and retrieve data efficiently in vector-based applications.


```python
#create a dummy database
data1 = [Document(page_content="langchain is a framework to create llm based application",metadata={"topic":"Langchain","difficultly":"easy"}),
        Document(page_content="langchain contains four important components langchain, langgraph, langsmith, langserve",metadata={"topic":"Langchain","difficultly":"medium"}),
        Document(page_content="ML stands for machine Learning",metadata={"topic":"machine Learning","difficultly":"easy"}),
        Document(page_content="Machine Learning having 3 types of machine Learning",metadata={"topic":"Machine Learning","difficultly":"easy"})
        ]
```

```python
["doc{}".format(y) for y in range(len(data1))]
```

```python
# Creating a Chroma database with custom IDs allows precise control over document identification, making it easier to manage, update, and retrieve data efficiently in vector-based applications.

db = Chroma.from_documents(documents = data1, embedding=emb,persist_directory="db4" , ids= ["doc{}".format(y) for y in range(len(data1))])
db.persist()
```
